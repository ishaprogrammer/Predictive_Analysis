```python
numeric_cols = df.select_dtypes(exclude=object).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
df.isna().sum()
```

```python
df.duplicated().any()
```

```python
df.head()
```

```python
df['Downtime'].unique()
```

```python
df.replace({'Downtime': {'Machine_Failure': 1, 'No_Machine_Failure': 0}}, inplace=True)
```

```python
df = df.drop(columns=['Date', 'Machine_ID', 'Assembly_Line_No'])
```

```python
df.head()
```

```python
df.columns
```

```python
df['Total_Pressure(bar)'] = df['Hydraulic_Pressure(bar)'] + df['Coolant_Pressure(bar)'] + df['Air_System_Pressure(bar)']
```

```python
df['Average_Temperature'] = (df['Coolant_Temperature'] + df['Hydraulic_Oil_Temperature(?C)'] + df['Spindle_Bearing_Temperature(?C)']) / 3
```

```python
df['Total_Vibration'] = df['Spindle_Vibration(?m)'] + df['Tool_Vibration(?m)']
```

```python
df.drop(columns=['Hydraulic_Pressure(bar)', 'Coolant_Pressure(bar)', 'Air_System_Pressure(bar)', 'Coolant_Temperature', 'Hydraulic_Oil_Temperature(?C)', 'Spindle_Bearing_Temperature(?C)', 'Spindle_Vibration(?m)', 'Tool_Vibration(?m)'], inplace=True)
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x='Downtime', data=df)
plt.title('Downtime Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df[['Spindle_Speed(RPM)', 'Voltage(volts)', 'Torque(Nm)', 'Cutting(kN)', 'Total_Pressure(bar)', 'Average_Temperature', 'Total_Vibration', 'Downtime']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
columns = ['Spindle_Speed(RPM)', 'Voltage(volts)', 'Torque(Nm)', 'Cutting(kN)', 'Total_Pressure(bar)', 'Average_Temperature', 'Total_Vibration']

for column in columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='Downtime', y=column, data=df)
    plt.title(f"{column} vs Downtime")
    plt.show()
```

```python
sns.pairplot(df[['Spindle_Speed(RPM)', 'Voltage(volts)', 'Torque(Nm)', 'Cutting(kN)', 'Total_Pressure(bar)', 'Average_Temperature', 'Total_Vibration', 'Downtime']], hue='Downtime')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# List of columns to scale
columns_to_scale = ['Spindle_Speed(RPM)', 'Voltage(volts)', 'Torque(Nm)', 'Cutting(kN)',
                    'Total_Pressure(bar)', 'Average_Temperature', 'Total_Vibration']

# Min-Max Scaler
scaler = MinMaxScaler()
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```
